```
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
from IPython.display import Image
plt.rc('figure', figsize=(10, 6))
import seaborn as sns
sns.set()
```

```
y=np.random.randn(30)
y
x=np.random.randn(30)
x
plt.plot(y)
plt.plot(x)
plt.legend(["random1", "random2"], )
```

```
plt.figure(figsize=(10,6))
y=np.random.randn(30)
y
x=np.random.randn(30)
x
z=np.random.randn(30)
z
#line Style and colot
plt.plot(y,linestyle="--", color="red")
plt.plot(x)
#marker
plt.plot(z, marker="o")
#Adding legend
plt.legend(["random1", "random2"], )

ax=plt.gca()
ax.set_xticks([0,1,7,12])

ax.set_ylim([-2,4])

ax.grid(True)
plt.xlabel("Index")
plt.ylabel("Random Int")
#annotating different obserations
plt.annotate(xy=[0,1], s="start")
plt.annotate(xy=[4,3], s="Highest")

plt.savefig('mysample.png')
```
